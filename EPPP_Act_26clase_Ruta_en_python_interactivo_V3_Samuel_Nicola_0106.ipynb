```python
# Clase Ruta con 3 atributos origen, destino, distancia
# Funciones leerdatos(), mostardatos()
# crear 2 objetos ruta1, ruta2

class Ruta:
  def __init__(self, origen, destino, distancia):
      self.origen = origen
      self.destino = destino
      self.distancia = distancia

  def leerdatos(self):
      self.origen = input("Ingrese el origen de la Ruta: ")
      self.destino = input("Ingrese el destino de la Ruta: ")
      self.distancia = input("Ingrese la distancia de la Ruta: ")
      return self.origen, self.destino, self.distancia

  def mostardatos(self):
      print("Origen: ", self.origen)
      print("Destino: ", self.destino)
      print("Distancia: ", self.distancia)
      return self.origen, self.destino, self.distancia
# fin de la clase ruta

print("ruta numero 1")
ruta1 = Ruta("","","") # creando el objeto ruta1
ruta1.leerdatos()
ruta1.mostardatos()
print("_*_*_*_*_*_*_*_*_*_*_*_*_*_")
print("ruta numero 2")
ruta2 = Ruta("","","") # creando el objeto ruta2
ruta2.leerdatos()
ruta2.mostardatos()
print("_*_*_*_*_*_*_*_*_*_*_*_*_*_")

print("programa elaborado por Samuel Nicola NL = 40 + 3 = 43")
```
